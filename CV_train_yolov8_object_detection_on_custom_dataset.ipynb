```python
import cv2
import time
from ultralytics import YOLO
from playsound import playsound
import threading
from plyer import notification  # Importing plyer for desktop notifications

#  Load the three different YOLOv8 trained models
model1 = YOLO(r"runs\detect\train\weights\best.pt")
model2 = YOLO(r"runs\detect\train2\weights\best.pt")
model3 = YOLO(r"runs\detect\train22\weights\best.pt")

#  Load video file
video_path = r"accident12.mp4"
cap = cv2.VideoCapture(video_path)

#  Output video setup
output_path = "detection_output.mp4"
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

#  Alert control
accident_alert_triggered = False
fire_alert_triggered = False

#  Start showing window immediately
cv2.namedWindow("Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Detection", 800, 600)

def play_alert():
    try:
        playsound("alert.mp3")
    except Exception as e:
        print(" Sound Error:", e)

def send_notification(message):
    # Send desktop notification
    notification.notify(
        title=" Incident Detected!",
        message=message,
        app_name="Fire & Accident Detection",
        timeout=25  # Notification will disappear after 10 seconds
    )

def handle_notification(name):
    # Send different notifications based on the type of incident detected
    if name == "fire":
        send_notification(" Fire detected! Please take action immediately!")
    elif name == "accident":
        send_notification(" Accident detected! Please respond ASAP!")
    else:
        send_notification(f"Alert: {name.capitalize()} detected!")

#  Main loop
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Inference
    results1 = model1(frame, verbose=False)
    results2 = model2(frame, verbose=False)
    results3 = model3(frame, verbose=False)

    all_results = results1 + results2 + results3

    for r in all_results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls = int(box.cls[0])
            name = model1.names.get(cls, 'Unknown').lower()
            label = f"{name} {conf:.2f}"

            # Draw detection box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

            # Alert only once and send notification
            if name in ["fire", "accident"] and not (accident_alert_triggered if name == "accident" else fire_alert_triggered):
                print(f" {name.capitalize()} Detected!")
                threading.Thread(target=play_alert, daemon=True).start()
                threading.Thread(target=handle_notification, args=(name,), daemon=True).start()

                # Set the respective flag based on detection type
                if name == "accident":
                    accident_alert_triggered = True
                elif name == "fire":
                    fire_alert_triggered = True

    # Display frame and save output
    cv2.imshow("Detection", frame)
    out.write(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything
cap.release()
out.release()
cv2.destroyAllWindows()
print(" Detection complete. Output saved as:", output_path)
```

```python
import cv2
import time
from ultralytics import YOLO
from playsound import playsound
import threading
from plyer import notification
import collections
import os

# Load the three different YOLOv8 trained models
model1 = YOLO(r"runs\detect\train\weights\best.pt")
model2 = YOLO(r"runs\detect\train2\weights\best.pt")
model3 = YOLO(r"runs\detect\train22\weights\best.pt")

# Load video file
video_path = r"accident12.mp4"
cap = cv2.VideoCapture(video_path)

# Output video setup
output_path = "detection_output.mp4"
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Alert control
accident_alert_triggered = False
fire_alert_triggered = False

# Start showing window immediately
cv2.namedWindow("Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Detection", 800, 600)

# Buffer for last N seconds of frames
BUFFER_SECONDS = 5
frame_buffer = collections.deque(maxlen=fps * BUFFER_SECONDS)

def play_alert():
    try:
        playsound("alert.mp3")
    except Exception as e:
        print(" Sound Error:", e)

def send_notification(message):
    notification.notify(
        title=" Incident Detected!",
        message=message,
        app_name="Fire & Accident Detection",
        timeout=25
    )

def save_incident_clip(buffer, incident_type, cap, fps, width, height, seconds_after=10):
    incident_clip_path = f"{incident_type}_incident_clip.mp4"
    clip_writer = cv2.VideoWriter(
        incident_clip_path, fourcc, fps, (width, height)
    )
    # Write buffered frames (before incident)
    for f in buffer:
        clip_writer.write(f)
    # Write frames after incident
    frames_to_write = int(fps * seconds_after)
    for _ in range(frames_to_write):
        success, frame = cap.read()
        if not success:
            break
        clip_writer.write(frame)
    clip_writer.release()
    return incident_clip_path

def handle_notification(name, clip_path=None):
    message = ""
    if name == "fire":
        message = " Fire detected! Please take action immediately!"
    elif name == "accident":
        message = " Accident detected! Please respond ASAP!"
    else:
        message = f"Alert: {name.capitalize()} detected!"
    if clip_path:
        message += f"\nVideo clip saved: {clip_path}"
    send_notification(message)
    if clip_path:
        os.startfile(clip_path)  # Windows only

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    frame_buffer.append(frame.copy())

    results1 = model1(frame, verbose=False)
    results2 = model2(frame, verbose=False)
    results3 = model3(frame, verbose=False)

    all_results = results1 + results2 + results3

    for r in all_results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls = int(box.cls[0])
            name = model1.names.get(cls, 'Unknown').lower()
            label = f"{name} {conf:.2f}"

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

            if name in ["fire", "accident"] and not (accident_alert_triggered if name == "accident" else fire_alert_triggered):
                print(f" {name.capitalize()} Detected!")
                threading.Thread(target=play_alert, daemon=True).start()
                # Save incident clip with 10 seconds after detection
                incident_clip_path = save_incident_clip(
                    frame_buffer, name, cap, fps, width, height, seconds_after=10
                )
                threading.Thread(target=handle_notification, args=(name, incident_clip_path), daemon=True).start()
                if name == "accident":
                    accident_alert_triggered = True
                elif name == "fire":
                    fire_alert_triggered = True

    cv2.imshow("Detection", frame)
    out.write(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
print(" Detection complete. Output saved as:", output_path)
```
